# Описание данных

* Sex - пол (M: мужской, F: женский, I: ребёнок)
* Length - длина корпуса (мм)
* Diameter - диаметр (мм), перпендикулярно длине
* Height - высота - с мясом в скорлупе (мм)
* Whole weight - общий вес (г)
* Shucked weight - вес очищенного мяса (г)
* Viscera weight - масса внутренних органов после кровотечения (г)
* Shell weight - масса скорлупы - после сушки (г)
* Rings - кольца - значения + 1.5 = возраст в годах (например, 4 = 5,5 лет)

## Заполнение пропущенных значений

Т.к. кол-во прощущенных значений в каждом из столбцов (Diameter, Whole weigth, Shell weight) меньше 10%, но больше 1%, то заполним пропуски медианными значениями по столбцу

```python
median_diameter = df['Diameter'].median()
df['Diameter'].fillna(median_diameter, inplace=True)

median_whole_weight = df['Whole weight'].median()
df['Whole weight'].fillna(median_whole_weight, inplace=True)

median_shell_weight = df['Shell weight'].median()
df['Shell weight'].fillna(median_shell_weight, inplace=True)
```

## Связь признаков с целевой переменной Rings

Проверим есть ли в данном столбце ошибочные значения

```python
df['Sex'].value_counts(normalize=True)
```

Есть ошибочное значение в виде того, что есть строчная буква f. Заменим ее на заглавную

Посмотрим, как пол влияет на кол-во колец

Как и ожидалось меньше всего колец будет у моллюска категории I, а наибольшее кол-во колец будет у категории F

### Length, Diameter, Height

```python
for col in ['Length', 'Diameter', 'Height']:
    plt.figure(figsize=(6,4))
    plt.title(f'{col}-Rings')
    plt.scatter(df[col], df.Rings);
    print(f"Коэф. корреляции между {col} и Rings: {np.corrcoef(df[col], df.Rings)[0][1]}")
```

Как мы видим, чем больше длина, диаметр и высота, тем больше колец. К тому же об этом свидетельствует коэф. корреляции

```python
for col in ['Whole weight', 'Shucked weight', 'Viscera weight', 'Shell weight']:
    plt.figure(figsize=(6,4))
    plt.title(f'{col}-Rings')
    plt.scatter(df[col], df.Rings);
    print(f"Коэф. корреляции между {col} и Rings: {np.corrcoef(df[col], df.Rings)[0][1]}")
```

Как и ожидалось, данные признаки хорошо коррелируют с целевым признаком, поэтому с увеличением любого из веса следует увеличение кол-ва колец

# Корреляции признаков

```python
corr = df.corr(numeric_only=True) # матрица корреляций

sns.heatmap(corr, cmap="Blues", annot=True)
```

Мы видим высокую корреляцию между признаками, из которой следует чем больше признак, тем больше колец

```python
from scipy.stats import f_oneway

Data = []

for c1 in df.columns:
    for c2 in df.columns:
        if df[c1].dtype == 'object' and df[c2].dtype != 'object':
            CategoryGroupLists = df.groupby(c1)[c2].apply(list)
            AnovaResults = f_oneway(*CategoryGroupLists)

            if AnovaResults[1] >= 0.05:
                Data.append({'Category' : c1, 'Numerical' : c2, 'Is correlated' : 'No'})
            else:
                Data.append({'Category' : c1, 'Numerical' : c2, 'Is correlated' : 'Yes'})

AnovaRes = pd.DataFrame.from_dict(Data)
AnovaRes
```

Также наблюдается корреляция между полом и числовыми признаками

# Анализ целевого признака

```python
df['Rings'].hist(bins=100, figsize=(10,8))
```

```python
from scipy.stats import normaltest

pvalue = normaltest(df['Rings'])
print("P-value", pvalue.pvalue)
print("Нормальное распределение данных: ", float(pvalue.pvalue) > 0.05)
```

Исходя из гистограммы и теста на нормальное распределение можно сделать вывод, что распределение целевого признака является ненормальным, как следствие могут возникнуть проблемы при обучении модели, а именно модель будет больше всего ошибаться в значениях, которые находятся по краям (в хвостах)

```python
import polars as pl
```

## Загрузка данных

```python
# Polars
%%time

df_pl = pl.read_csv("abalone_big.csv")
```

```python
# Pandas
%%time

df = pd.read_csv("abalone_big.csv")
```

Как мы видим загрузка данных у polars немного дольше, чем у pandas, но эта разница совсем маленькая

## Фильтрация данных

```python
# Pandas
%%time

df[['Sex', 'Length']]
```

```python
# Polars
%%time

df_pl[['Sex', 'Length']]
```

```python
# Pandas
%%time

df.query('Length > 0.03')
```

```python
%%time
# Polars

df_pl.filter(pl.col('Length') > 0.03)
```

Как мы видим Polars отрабатывает чуть быстрее Pandas

## Агрегация данных

```python
%%time

# Pandas
df.groupby('Sex').agg({'Diameter': 'mean', 'Height':'mean'})
```

```python
%%time

df_pl.group_by('Sex').agg([pl.mean('Diameter'), pl.mean('Height')])
```

```python
%%time

q = (
    df_pl
    .lazy()
    .group_by(by='Sex')
    .agg(
        [
            pl.col('Diameter').mean(),
            pl.col('Height').mean()
        ]
    )
)
```

```python
%%time

q.collect()
```

Получилось так, что при обычной группировки Polars отрабатывает быстрее, а с использованием lazy-метода в разы быстрее, чем обычная группировка у Pandas
